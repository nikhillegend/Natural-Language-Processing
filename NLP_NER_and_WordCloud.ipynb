```python
import spacy
from spacy import displacy
import pandas as pd
```

```python
df = pd.read_csv('/content/apple.txt',on_bad_lines='skip')
#to load the txt file with read_csv() keep on_bad_lines='skip
```

```python
df
```

```python
df['x'] #x is a column name
```

```python
# We want to join all these sentence
```

```python
nlp = spacy.load('en_core_web_sm')#model name is en_core_web_sm, nlp will now have entire NER model
```

```python
nlp.pipe_names
```

```python
#We wabt ner (name entity recognition)
```

```python
doc = nlp('Mr.Elon is going to acquire Twitter Inc for $45 billion in 2023')
```

```python
doc.ents #will display all entities
```

```python
for i in doc.ents:
  print(i, '|', i.label_)
```

```python
doc1 = nlp('Elon is going to acquire Twitter Inc for $45 billion in 2023')
```

```python
doc1.ents
```

```python
for i in doc1.ents:
  print(i, ':-', i.label_)
```

```python
#Spacy is not 100 precise
```

```python
#to display graphical representation for tags
displacy.render(doc, style='ent',jupyter=True)
```

```python
displacy.render(doc1, style='ent', jupyter=True)
```

```python
#joining all sentences of text
txt = ' '.join(df['x'])
txt
```

```python
x = nlp(txt)
```

```python
displacy.render(x, style='ent', jupyter=True)
```

```python
for i in x:
  print(i, i.pos_)
```

```python
wc = WordCloud(background_color='black', stopwords=stop).generate(txt)
plt.imshow(wc)
```
