```python
 #Sentence Tokenization
 sent = nltk.sent_tokenize(para)
```

```python
len(sent)
```

```python
sent[0] #Shows first document
```

```python
 #Text cleaning - removing unnecessary things - punctuatuions marks, symbols, emojis, etc. using sub()
 #Text normalization - converting each word in lower case
 corpus = []

 for i in range(len(sent)):
    txt = re.sub('[^a-zA-Z]',' ', sent[i])# expect a-zA-Z remove everything from each sentence or document
    txt = txt.lower()
    corpus.append(txt)
```

```python
corpus
```

```python
#now we are going to perform stemming and lemmatization
#Stemming - before applying stemming on document, trying to know stemming works
stemmer = PorterStemmer()
```

```python
stemmer.stem('goes')
```

```python
stemmer.stem('history')
```

```python
stemmer.stem('finally')
```

```python
for i in corpus:
    words = nltk.word_tokenize(i)
    print(words)
    #we get seperate list for each sentence
```

```python
#Performing Tokenization, stemming and stop word removal

for i in corpus:
    words = nltk.word_tokenize(i) # for each sentence in corpus perform word tokenization
    for i in words:
      if i not in set(stopwords.words('english')):
        print(stemmer.stem(i))
```

```python
# Lemmantization
lemma = WordNetLemmatizer()
```

```python
lemma.lemmatize('google')
```

```python
lemma.lemmatize('historical')
```

```python
lemma.lemmatize('coming')
```

```python
for i in corpus:
    words = nltk.word_tokenize(i)
    for i in words:
      if i not in set(stopwords.words('english')):
        print(lemma.lemmatize(i))
```
